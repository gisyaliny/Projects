# ArcGIS Online and Story Map

## ArcGIS Online

- [Learning Resources](https://www.esri.com/en-us/arcgis/products/arcgis-online/resources)

`ArcGIS Online` is a cloud-based mapping and analysis platform developed by Esri that enables users to create, share, and manage geographic information, maps, and applications.

It provides a comprehensive set of tools and services for collecting, visualizing, analyzing, and sharing spatial data, making it an essential platform for GIS professionals, organizations, and individuals across various industries.

## [Story Map](https://storymaps.arcgis.com/stories)

`ArcGIS StoryMaps` is a dynamic web application developed by Esri that enables users to create engaging, interactive, and multimedia-rich stories using maps, images, videos, and text.

It provides a platform for storytelling with spatial data, allowing users to combine geographic information with narrative content to convey information, educate, and inspire their audience.
